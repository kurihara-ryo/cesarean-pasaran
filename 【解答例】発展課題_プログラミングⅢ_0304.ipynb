課題1

accuracyなどの精度指標がクラス確率の閾値に依存することを確認してみましょう。

10個のデータに対する2クラス分類問題を考えます。以下の`y_test`は真のクラスを表しており、`y_test_prob`は分類モデルによって予測されたクラス確率を表しています。`y_test_prob`の0列目がクラス0の確率、1列目がクラス1の確率です。通常、クラス確率の閾値には0.5を使うので、クラス1の確率が0.5より大きいデータをクラス1、それ以外のデータをクラス0に分類します。この結果が`y_test_pred`に入っています。

閾値を0.7にした場合のクラス分類の結果`y_test_pred_07`と、これに対するaccuracyを計算して表示してください。また、閾値を0.2にした場合のクラス分類の結果`y_test_pred_02`とaccuracyを計算して表示してください。

```python
import numpy as np
from sklearn.metrics import accuracy_score

# 真のクラス
y_test = np.array([0, 1, 1, 0, 0, 0, 0, 1, 0, 0])

# クラス確率
y_test_prob = np.array([0.62545988, 0.37454012,
                        0.04928569, 0.95071431,
                        0.26800606, 0.73199394,
                        0.40134152, 0.59865848,
                        0.84398136, 0.15601864,
                        0.84400548, 0.15599452,
                        0.94191639, 0.05808361,
                        0.13382385, 0.86617615,
                        0.39888499, 0.60111501,
                        0.29192742, 0.70807258])
y_test_prob = y_test_prob.reshape(10, 2)

# クラス確率の閾値を0.5にした場合の予測クラス
y_test_pred = np.zeros(y_test_prob.shape[0])
for i in range(y_test_prob.shape[0]):
    if y_test_prob[i, 1] > 0.5:
        y_test_pred[i] = 1
    else:
        y_test_pred[i] = 0
print('y_test_pred with threshold=0.5\n{}'.format(y_test_pred))
# 閾値0.5でのaccuracy
print('accuracy with threshold=0.5\n{}'.format(accuracy_score(y_true=y_test, y_pred=y_test_pred)))

# クラス確率の閾値を0.7にした場合の予測クラス
y_test_pred_07 = np.zeros(y_test_prob.shape[0])
for i in range(y_test_prob.shape[0]):
    if y_test_prob[i, 1] > 0.7:
        y_test_pred_07[i] = 1
    else:
        y_test_pred_07[i] = 0
print('y_test_pred with threshold=0.7\n{}'.format(y_test_pred_07))
# 閾値0.7でのaccuracy
print('accuracy with threshold=0.7\n{}'.format(accuracy_score(y_true=y_test, y_pred=y_test_pred_07)))

# クラス確率の閾値を0.2にした場合の予測クラス
y_test_pred_02 = np.zeros(y_test_prob.shape[0])
for i in range(y_test_prob.shape[0]):
    if y_test_prob[i, 1] > 0.2:
        y_test_pred_02[i] = 1
    else:
        y_test_pred_02[i] = 0
print('y_test_pred with threshold=0.2\n{}'.format(y_test_pred_02))
# 閾値0.2でのaccuracy
print('accuracy with threshold=0.2\n{}'.format(accuracy_score(y_true=y_test, y_pred=y_test_pred_02)))
```

課題2

演習課題では、モデルの性能評価を目的として、データを学習データとテストデータに分割して、学習データに対してハイパラの調整とモデルの学習を行い、テストデータに対する予測精度を評価しました。一方、機械学習を実問題に適用する際には、データ全体を用いてハイパラ調整とモデル学習を行い、ラベルの付いていない（正解の分からない）新しいデータに対して予測を行う必要があります。このような解析を体験してみましょう。

heart_failure_clinical_records.csvを用いて、演習課題のような性能評価を様々なモデルについて行った結果、このデータにおいてはランダムフォレストが最も良いモデルであると結論を出したと仮定します。その後、病院から新しい患者10名の臨床情報heart_failure_clinical_records_new.csvを受け取りました。これらの患者は存命中であるため、csvファイル内にDEATH_EVENTの列は存在しません。治療の方針や優先度を決めるために、機械学習で各患者の死亡確率を予測したいとします。

以下の手順に沿って、各患者の死亡確率を計算し、死亡確率の最も高い患者を同定してください。

1.   heart_failure_clinical_records.csvの全てのデータを使って、ランダムフォレストのハイパラ調整とモデル学習を行ってください。乱数シード、ハイパラの探索範囲、CVの分割数については、コード内の指示にしたがってください。精度指標はROC AUCとします。
2.   1.で作成したモデルを使って、heart_failure_clinical_records_new.csvの各患者に対してクラス確率を計算して表示してください。
3.   死亡確率（クラス1の確率）が最も高いのはデータフレーム内の何番目の患者か、コメントの解答欄に書いてください。一番上の患者を0番目とします。


```python
import numpy as np
import pandas as pd
import gdown

# 演習課題と同じデータ
url='https://drive.google.com/uc?export=download&id=170BvFfMQJBI8jueqdMXuJiB910bbKSiN'
gdown.download(url, 'heart_failure_clinical_records.csv', quiet=False)
df = pd.read_csv('heart_failure_clinical_records.csv')

# 新しい患者のデータ。DEATH_EVENTの列は存在しない。
url='https://drive.google.com/uc?export=download&id=1foR7RCoqlVva_rRuwBvpIo7OVzA2uVQ5'
gdown.download(url, 'heart_failure_clinical_records_new.csv', quiet=False)
df_new = pd.read_csv('heart_failure_clinical_records_new.csv')
display(df_new)

# 必要なモジュール一式を用意
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import GridSearchCV

# 乱数シードには以下の値を用いてください
seed = 4649

# ハイパラの探索範囲は以下のようにしてください
param_grid = {'n_estimators': [10, 20, 50, 100]}


#### ここからコードを書いてください

# 1) heart_failure_clinical_records.csvの全てのデータを使って、
# ハイパラ調整とモデルの学習を行う。

# 1-1) データフレームをX,yに分ける
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
print('X shape {}'.format(X.shape))
print('y shape {}'.format(y.shape))

# 1-2) Xの前処理
scaler = StandardScaler()
X = scaler.fit_transform(X)
# ここで作ったscalerは新しい患者データの前処理にも使用

# 1-3) モデルのインスタンスを作成
model = RandomForestClassifier(random_state=seed)

# 1-4) CVの方法を指定
# 分割数は5としてください
cv = KFold(n_splits=5, shuffle=True, random_state=seed)

# 1-5) グリッドサーチの方法を指定
# 精度指標にはROC AUCを使用してください
gs = GridSearchCV(estimator=model, param_grid=param_grid, cv=cv, scoring='roc_auc')

# 1-6) ROC AUCの最も良いハイパラを選択
gs.fit(X=X, y=y)
print('best hyperparameter: {}'.format(gs.best_params_))
print('best ROC AUC: {}\n'.format(gs.best_score_))


# 2) 1)で作成したモデルを使って、
# heart_failure_clinical_records_new.csvの各患者に対してクラス確率を計算

# 2-1) DEATH_EVENTは存在しないため、X,yに分ける必要はない
# データフレーム全体をX_newとする
X_new = df_new

# 2-2) データの前処理
# Xと同じ基準で変換するために、Xに使用したのと同じインスタンスを流用
X_new = scaler.transform(X_new)

# 2-3) クラス確率の計算
y_new_prob = gs.predict_proba(X_new)
print('Class probability for new patients:\n{}'.format(y_new_prob))


# 3) 死亡確率の最も高い患者を同定
print('Patient with highest death probability:\n{}'.format(y_new_prob[:, 1].argmax()))

# 解答欄：死亡確率の最も高い患者は？
# 3番目の患者
```

課題3

プロモータとは、遺伝子の上流部分にあり、その遺伝子の転写を活性化する機能をもった領域のことを指します。プロモータのDNA配列の違いによって、遺伝子の転写活性が変化することが知られています。近年では、機械学習を用いてプロモータのDNA配列から転写活性を予測することで、人工的な高活性プロモータをデザインする研究が行われています。

promoter_activity.csvには、様々な人工プロモータのDNA配列とその転写活性が記載されています。このデータを用いて、DNA配列から転写活性を予測するランダムフォレストの回帰モデルを作成し、予測精度を評価してください。

*   入力Xには、データフレーム内のpromoter_sequenceからone-hot encodingを計算して使用してください。
*   予測対象yには、データフレーム内のtranscription_activityを用いてください。
*   乱数シード、学習データとテストデータの分割比率、ハイパラの探索範囲、CVの分割数については、コード内の指示にしたがってください。精度指標はR2とします。

データの出典：[LaFleur et al. Nat Commun. 2022.](https://doi.org/10.1038/s41467-022-32829-5) より一部抜粋。

```python
# データの確認
# このセルは実行するだけ
import gdown
import pandas as pd
import numpy as np

url='https://drive.google.com/uc?export=download&id=1OjDryc69PuhvsjK5B2wtmxwHSchU-dXN'
gdown.download(url, 'promoter_activity.csv', quiet=False)
df = pd.read_csv('promoter_activity.csv')
display(df)
```

```python
# ここから課題

# 1) Xの作成
# df内のpromoter_sequenceからone-hot encodingを計算

# ヒント1：
# df内の配列はどれも長さ73なので、one-hot encodingのshapeは(73, 4)になる。
# これを特徴ベクトルとして使用するために、
# flatten()などのメソッドを使って1次元アレイに変換する。
# 特徴ベクトルの長さは73 x 4 = 292になる。
Base2Index = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
def onehot_encoding(s):
    res = np.zeros((len(s), 4))
    for i in range(len(s)):
        res[i, Base2Index[s[i]]] = 1
    return res.flatten() # 1次元アレイに変換

# ヒント2：
# Xはshapeが(データ数, 292)の行列になる。
# one-hot encodingは標準化せずに用いるのが一般的なので、
# StandardScaler()は使用せずにそのまま用いてよい。
X = np.zeros([df.shape[0], 292])
for i in range(df.shape[0]):
    X[i, :] = onehot_encoding(df['promoter_sequence'][i])
print(X.shape)

# 2) yの作成
# df内のtranscription_activityをそのまま使用
y = df['transcription_activity']


# 3) モデルの作成と評価

# 必要なモジュール一式を用意
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import GridSearchCV

# 乱数シードには以下の値を用いてください
seed = 1234

# X,yをそれぞれ学習データとテストデータに分ける
# 分割比率は、train:test = 7:3としてください
X_train,  X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=seed)
print('X_train shape {}'.format(X_train.shape))
print('X_test shape {}'.format(X_test.shape))
print('y_train shape {}'.format(y_train.shape))
print('y_test shape {}'.format(y_test.shape))

# モデルのインスタンスを作成
model = RandomForestRegressor(random_state=seed)
# ハイパラの探索範囲は以下のようにしてください
param_grid = {'n_estimators': [10, 20, 50, 100]}

# CVの方法を指定
# 分割数は5としてください
cv = KFold(n_splits=5, shuffle=True, random_state=seed)

# グリッドサーチの方法を指定
# 精度指標にはR2を使用してください
gs = GridSearchCV(estimator=model, param_grid=param_grid, cv=cv, scoring='r2')

# グリッドサーチを実行
# R2の最も良いハイパラを選択
gs.fit(X=X_train, y=y_train)
print('best hyperparameter: {}'.format(gs.best_params_))
print('best R2: {}'.format(gs.best_score_))

# テストデータに対する予測精度を評価
y_test_pred = gs.predict(X_test) # 予測値

r2_test = r2_score(y_true=y_test, y_pred=y_test_pred)
print('R2 on test data: {}'.format(r2_test))
```
