【課題】リストscoresは、ある科目における学生の成績の点数を表しています。各点数をA,B,C,不合格の評定に変換して、リストgradesに格納して表示してください。評定の基準は北里大と同じとします（80点以上：A、79〜70点：B、69〜60点：C、59点以下：不合格）。

※「プログラミングⅡ」第1,2回の内容

```python
scores = [95, 68, 72, 83, 40, 61, 79, 100]
grades = []

for s in scores:
    if s >= 80:
        grades.append('A')
    elif s >= 70:
        grades.append('B')
    elif s >= 60:
        grades.append('C')
    else:
        grades.append('不合格')

print(grades)
```

【課題】辞書menuは、ある飲食店における料理とその値段を表しています。各料理の値段を20％値上げするようにmenuを変更してください。その後、1000円以下の料理を辞書menu_cheapに格納して表示してください。

※「プログラミングⅡ」第1,2回の内容

```python
menu = {'パスタ' : 1200, 'ラーメン' : 800, 'カレー' : 900, 'サラダ' : 500}
menu_cheap = {}

# menuの各料理を20％値上げ
for k,v in menu.items():
    menu[k] = 1.2 * v

# 1000円以下の料理をmenu_cheapに格納
for k,v in menu.items():
    if menu[k] <= 1000:
        menu_cheap[k] = v

print(menu_cheap)
```

【課題】自然数のリストnumsを引数として、リスト内の奇数の個数を戻り値とする関数count_odds()を書いてください。

※「プログラミングⅡ」第1,2回の内容

```python
def count_odds(nums):
    count = 0
    for n in nums:
        if n % 2 == 1: # 奇数
            count += 1
    return count

# 動作確認用：ここから下は変更しない
print(count_odds([10, 5, 3, 8, 7, 2])) # 3個
print(count_odds([1, 3, 5, 7, 9, 11])) # 6個
```

【課題】人名のリストnames1において、それぞれの人名は'Shibasaburo Kitasato'のように「名前 名字」の順に並んでいます。これを'KITASATO Shibasaburo'のように「名字 名前」の順にして、さらに名字を全て大文字に変換して、リストnames2に格納して表示してください。

※「プログラミングⅡ」第3,4回の内容

```python
# ヒント：
# split()で名前と名字を分割してリストにする
# スライスのstepを使ってリストの順番を逆にする
# upper()で名字を大文字にする
# join()で名字と名前を連結する

names1 = ['Shibasaburo Kitasato', 'Kotaro Oka', 'Shohei Ohtani', 'Ichiro Suzuki']
names2 = []

for n1 in names1:
    nm = n1.split() # 名前と名字を分割
    mn = nm[::-1] # リストの順番を逆にする
    mn[0] = mn[0].upper() # 名字を大文字にする
    n2 = ' '.join(mn) # 名字と名前を連結
    names2.append(n2)

print(names2)
```

【課題】「人工知能・機械学習入門」で学んだRBFカーネルをPythonで書いてみましょう。

$$
k(\mathbf{x}, \mathbf{y}) = \exp \left( -\gamma \| \mathbf{x} - \mathbf{y} \|^{\,2} \right)
$$

numpyアレイとして表されたベクトルx,yを引数として、RBFカーネルの計算結果を戻り値とする関数rbf_kernel()を書いてください。$\gamma$は0.1として、アレイのベクトル演算を利用して1行で書いてください。

※「プログラミングⅡ」第5,6回の内容

```python
import numpy as np

def rbf_kernel(x, y):
    return np.exp(-0.1 * np.sum(x - y) ** 2)

# 動作確認用：ここから下は変更しない
v = np.array([1, 2])
w = np.array([3, 4])
print(rbf_kernel(v, w)) # 0.449...
v = np.array([3, -5])
w = np.array([-2, 2])
print(rbf_kernel(v, w)) # 0.000611...
```

【課題】自動販売機を表すクラスVendingMachineを定義してください。

以下のメンバ変数を持たせてください。

*  商品名をキー、価格を値とする辞書`item2price`
*  商品名をキー、自販機内の在庫数を値とする辞書`item2num`
*  自販機内に溜まっているお金の量を表す変数`amount`

以下のメソッドを持たせてください。

*  コンストラクタ: `amount`を指定された値で初期化する。デフォルト値は`amount=0`とする。`item2price`, `item2num`は引数によらず空の辞書で初期化する。
* `set_price(self, item, price)`: `item`の価格を`price`円に設定するメソッド。
* `add_item(self, item, num)`: `item`の在庫を`num`個、自販機に追加するメソッド。
* `sell(self, item)`: `item`を1つ売る際の処理を行うメソッド。`item`の在庫を1つ減らして`amount`を価格の分だけ増やす。ただし、`item`が`item2price`や`item2num`にキーとして存在しないなら、何もせずに終了する。また、もし在庫がないなら「売り切れです」と表示して終了する。

※「プログラミングⅡ」第19,20回の内容

```python
class VendingMachine:
    def __init__(self, amount=0):
        self.item2price = {}
        self.item2num = {}
        self.amount = amount

    def set_price(self, item, price):
        self.item2price[item] = price

    def add_item(self, item, num):
        if item in self.item2num:
            self.item2num[item] += num
        else:
            self.item2num[item] = num

    def sell(self, item):
        if (not item in self.item2price) or (not item in self.item2num):
            return
        if self.item2num[item] > 0:
            self.item2num[item] -= 1
            self.amount += self.item2price[item]
        else:
            print('{}は売り切れです'.format(item))


# 動作確認用：ここから下は変更しない
vm = VendingMachine()
vm.set_price('コーラ', 150) # コーラは150円
vm.add_item('コーラ', 10) # コーラを10個補充
vm.set_price('ビール', 300) # ビールは300円
vm.add_item('ビール', 1) # ビールを1個補充
print(vm.item2num)
vm.sell('コーラ')
print(vm.item2num)
vm.sell('ビール')
print(vm.item2num)
vm.sell('ビール') # 売り切れ
vm.sell('コーヒー')
```
