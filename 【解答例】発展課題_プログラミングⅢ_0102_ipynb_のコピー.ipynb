課題1

タンパク質のアミノ酸配列`s`を引数として、one-hot encodingを戻り値とする関数`onehot_encoding_aa()`を書いてください。one-hot encodingはshapeが`(len(s), 20)`の2次元アレイとし、どのアミノ酸が何列目に対応するかは辞書`AA2Index`を参照してください。

この関数を使って、下記の配列`seq`のone-hot encodingを表示してください。



```python
import numpy as np

AA2Index = {
    'A':  0, 'R':  1, 'N':  2,  'D': 3, 'C':  4, 'Q':  5, 'E':  6, 'G':  7, 'H':  8, 'I':  9,
    'L': 10, 'K': 11, 'M': 12, 'F': 13, 'P': 14, 'S': 15, 'T': 16, 'W': 17, 'Y': 18, 'V': 19
}

seq = 'MGLSDGEWQLVLNVW'

def onehot_encoding_aa(s):
    # まず、0のみからなるlen(s)行20列の行列を作る。
    res = np.zeros((len(s), 20))
    # 次に、sを1文字ずつ見ながら、行列の該当する場所に1を代入していく。
    for i in range(len(s)):
        res[i, AA2Index[s[i]]] = 1
    return res

# 簡単な例で動作確認
print('one-hot encoding for AN:')
print(onehot_encoding_aa('AN'), '\n')

# seqのone-hot encodingを表示
print('one-hot encoding for {}:'.format(seq))
print(onehot_encoding_aa(seq))
```

課題2

カーネル行列とは、データの全てのペアに対してカーネル関数を計算して、それらの値を行列に格納したものです。`n`個のデータがあるとき、カーネル行列`K`の`i`行`j`列の要素は、`i`番目のデータと`j`番目のデータのカーネル関数の値になります。

$$
K_{i,j} = k(\mathbf{x}_{i}, \mathbf{x}_{j})
$$

下記の2次元アレイ`X`は、5個の3次元データを表しています。この`X`と演習課題で書いた関数`rbf_kernel()`を使って、カーネル行列を計算して表示してください。

```python
import numpy as np

# RBFカーネル
def rbf_kernel(x, y):
    # 演習課題で書いたものをコピペする
    return np.exp(-0.1 * np.sum((x - y) ** 2))

# 5個の3次元データ
X = np.array([0.50, -0.14,  0.65,
              1.52, -0.23, -0.23,
              1.58,  0.77, -0.47,
              0.54, -0.46, -0.47,
              0.24, -1.91, -1.72])
X = X.reshape(5, 3)

# カーネル行列を計算
K = np.zeros((X.shape[0], X.shape[0]))
for i in range(X.shape[0]):
    # 対称行列なので上三角部分だけ計算してコピーしてもいい
    for j in range(i, X.shape[0]):
        K[i][j] = rbf_kernel(X[i, :], X[j, :])
        if i != j:
            K[j][i] = K[i][j]

# カーネル行列を表示
print('Kernel matrix K:')
print(K)
```

課題3

辞書`Codon2AA`は、コドンとアミノ酸の対応関係（遺伝暗号）を表しています。遺伝暗号表において、同じアミノ酸をコードしているコドンを同義コドンと呼びます。例えば、TTTとTTCはどちらもアミノ酸Fをコードしているため同義コドンの関係にあります。

下記の配列seqに対して、同じアミノ酸配列をコードする塩基配列が何通り存在するか計算してください。



```python
Codon2AA = {
    "TTT":"F","TTC":"F","TTA":"L","TTG":"L","CTT":"L","CTC":"L","CTA":"L","CTG":"L",
    "ATT":"I","ATC":"I","ATA":"I","ATG":"M","GTT":"V","GTC":"V","GTA":"V","GTG":"V",
    "TCT":"S","TCC":"S","TCA":"S","TCG":"S","AGT":"S","AGC":"S","CCT":"P","CCC":"P",
    "CCA":"P","CCG":"P","ACT":"T","ACC":"T","ACA":"T","ACG":"T","GCT":"A","GCC":"A",
    "GCA":"A","GCG":"A","TAT":"Y","TAC":"Y","CAT":"H","CAC":"H","CAA":"Q","CAG":"Q",
    "AAT":"N","AAC":"N","AAA":"K","AAG":"K","GAT":"D","GAC":"D","GAA":"E","GAG":"E",
    "TGT":"C","TGC":"C","TGG":"W","CGT":"R","CGC":"R","CGA":"R","CGG":"R","AGA":"R",
    "AGG":"R","GGT":"G","GGC":"G","GGA":"G","GGG":"G","TAA":"*","TAG":"*","TGA":"*"
}

seq = 'MGLSDGEWQLVLNVW'

# 考え方
# 例として、2文字のアミノ酸配列ANを考える。
# 遺伝暗号表を見ると、Aには同義コドンが4種類、Nには同義コドンが2種類存在する。
# そのため、ANをコードする塩基配列は4 * 2 = 8種類存在する。
#
# つまり、配列の各文字について、同義コドンの数を掛けあわせていけばいい。
# アミノ酸をキー、同義コドンの数を値とする辞書があると便利そうだ。

# アミノ酸をキー、同義コドンの数を値とする辞書
aa2n = {}
for k,v in Codon2AA.items():
    if v not in aa2n:
        aa2n[v] = 1
    else:
        aa2n[v] += 1

# seqをコードする塩基配列の数を計算
n_synonymous = 1
for i in range(len(seq)):
    n_synonymous *= aa2n[seq[i]]

print(n_synonymous)
```

課題4

work_record.csvをデータフレームとして読み込み、Aさん、Bさん、Cさんそれぞれについて、1年間の給料の合計金額を計算してください。給料は時給制であり、以下のルールで決まるものとします。

*  平日は時給2,500円。ただし、1日8時間以上勤務した場合、8時間を超えた分については時給3,000円。
*  土曜日曜は時給2,800円。ただし、1日8時間以上勤務した場合、8時間を超えた分については時給3,500円。

例えば、平日に10時間勤務すると、その日の給料は`2,500*8 + 3000*2 = 26,000`円になります。

```python
import gdown
import pandas as pd
import numpy as np

# ここは演習課題と同じ
url='https://drive.google.com/uc?export=download&id=1an_a2CVQzX9gyeQ2_IxoU2EYjieC-mIC'
gdown.download(url, 'work_record.csv', quiet=False)
df = pd.read_csv('work_record.csv')
display(df)
print('\n')

# 考え方の例
# 1) まず土曜日曜について給料の合計金額を計算する
# 1-1) query()を使って土曜日曜のデータを抽出
# 1-2) iloc[]を使って勤務時間の部分を切り出す
# 1-3) apply()を使って、3人それぞれについて給料の和を計算
# apply()に入れる関数は自分で書く必要がある
# 労働時間の入ったリストを引数として、給料の和を計算する関数が必要

# 2) 1と同様に平日の給料の合計金額を計算する

# 3) 1と2の結果を合計する

# 勤務時間のリストから給料の和を計算する関数、土曜日曜バージョン
def salary_sum_weekend(hours):
    s = 0
    for h in hours:
        if h < 8:
            s += 2800 * h
        else:
            s += 2800 * 8 + 3500 * (h - 8)
    return s

# 勤務時間のリストから給料の和を計算する関数、平日バージョン
def salary_sum_weekday(hours):
    s = 0
    for h in hours:
        if h < 8:
            s += 2500 * h
        else:
            s += 2500 * 8 + 3000 * (h - 8)
    return s

print('土曜日曜の給料の合計金額')
df_weekend = df.query("曜日 == '土' or 曜日 == '日'")
weekend = df_weekend.iloc[:, 4:].apply(salary_sum_weekend, axis=0)
weekend.index = ['Aさん', 'Bさん', 'Cさん']
display(weekend)
print('\n')

print('平日の給料の合計金額')
df_weekday = df.query("曜日 != '土' and 曜日 != '日'")
weekday = df_weekday.iloc[:, 4:].apply(salary_sum_weekday, axis=0)
weekday.index = ['Aさん', 'Bさん', 'Cさん']
display(weekday)
print('\n')

print('給料の合計金額')
display(weekday + weekend)
```
